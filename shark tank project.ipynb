```python
# Brand Names in which 2-4, Sharks have invested
two_sharks=st[st["total_sharks_invested"]==2]
three_sharks=st[st["total_sharks_invested"]==3]
four_sharks=st[st["total_sharks_invested"]==4]
```

```python
total_sharks=pd.concat([two_sharks,three_sharks,four_sharks])
total_sharks.reset_index(drop=True, inplace=True)
```

```python
fig=px.bar(total_sharks,x="brand_name",y="deal_amount",title="Investments in Brands made by 2-4 sharks",color="pitcher_ask_amount")
fig.update_layout(width=800,height=500)
fig.show()
```

```python
min=st.groupby("episode_number").min().reset_index()
```

```python
max=st.groupby("episode_number").max().reset_index()
```

```python
# Episode wise minimum and maximum of sharks invested?
plt.figure(figsize=(10,6))
sns.barplot(max,x="episode_number",y="total_sharks_invested",color="cyan")
sns.barplot(min,x="episode_number",y="total_sharks_invested",color="violet")
plt.show()
```

```python
# Minimum and Maximum sharks invested in each brand
brand_wise_invest=st.groupby("brand_name")[["deal_amount","amount_per_shark","ashneer_deal","anupam_deal",
                                            "aman_deal","namita_deal","vineeta_deal","peyush_deal","ghazal_deal"]].sum().reset_index()
```

```python
brand_wise_invest.sort_values(by='deal_amount', ascending=False, inplace=True)
```

```python
plt.figure(figsize=(16,6))
sns.barplot(brand_wise_invest,x="brand_name",y="deal_amount",palette="muted")
plt.xticks(rotation=45, ha='right', fontsize=6,fontweight='bold')
plt.title("maximum to minimum invested by sharks in each brand")
plt.show()
```

```python
# Brand name where deal equity is between 0-50
deal_=st[(st['deal_equity'] > 0) & (st['deal_equity'] < 50)]
# barplot
fig=px.bar(deal_,x="brand_name",y="equity_per_shark")
fig.show()
```
